```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('Pharmacies.csv')

x_column = df['ID']  
y_column = df['FID']  


sample_size = 1000 
df_sampled = df.sample(n=sample_size, random_state=1)


plt.figure(figsize=(10, 8))
sns.scatterplot(data=df_sampled, x='ID', y='FID')


plt.title(f"Scatter Plot of FID vs ID (Sampled)")
plt.xlabel('ID')
plt.ylabel('FID')


plt.show()
```
